```python
def mochila(weights, values, capacity, n):
    if n == 0 or capacity == 0: #T(n) = 1
        return 0 #T(n) = 1

    if weights[n-1] > capacity:
        return mochila(weights, values, capacity, n-1) #T(n) = n

    else:
        option1 = mochila(weights, values, capacity, n-1) #T(n) = 1
        option2 = values[n-1] + mochila(weights, values, capacity - weights[n-1], n-1) #T(n) = n
        return max(option1, option2)

weights = [10, 20, 30, 40] #T(n) = 1
values = [3, 4, 5, 6] #T(n) = 1
capacity = 5
n = len(weights) #T(n) = 1

#T(n) = 7 + 2n
#O(n)

resultado = mochila(weights, values, capacity, n)
print(f"El máximo valor que se puede obtener es: {resultado}")
```
